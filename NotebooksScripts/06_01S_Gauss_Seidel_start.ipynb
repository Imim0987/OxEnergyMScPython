## Gauss Seidel Starter - I


```python
import numpy as np
import matplotlib.pyplot as plt
eps = 1e-6
maxIter = 100
```


Find x using GS given
$$x^2 + 3x + 1 = 0$$

```python
xold = 1
e = 1

count=0
while (e > eps)&(count<maxIter):
    count += 1
    # Insert equations here:
    xnew = ((xold**2+1)/3)*-1  
    #
    e = np.absolute(xnew - xold)
    xold = xnew
if (count == maxIter):
    print(f"WARNING: Did not converge. e = {e:.4e}")

print(f"Answer = {xnew}. Interations = {count}")
```

## Gauss Seidel Starter - II

Find x using GS given
$$x^3-6x^2+9x-3= 0$$

```python
xold = 1
e = 1
count=0
while (e > eps)&(count<maxIter):
    count += 1
    # Insert equations here:
    xnew = -1*((xold**3-6*(xold)**2-3)/9)
    #
    e = np.absolute(xnew - xold)
    xold = xnew
if (count == maxIter):
    print(f"WARNING: Did not converge. e = {e:.4e}")

print(f"Answer = {xnew}. Interations = {count}")
```

## Gauss Seidel Starter - III

Find x using GS given
$$x^3-6x^2+9x-3= 0$$
$$y^2 -x = 0$$

```python
import numpy as np
from math import sqrt

import matplotlib.pyplot as plt

xold = 2
yold = 2
e=1
eps = 1e-6
maxIter = 100  

count=0
while (e > eps)&(count<maxIter):
    # Insert equations here:
    xnew = -1*((xold**3-6*(xold)**2-3)/9)
    ynew = sqrt(xold)
    #
    e = max(np.absolute(xnew - xold), np.absolute(ynew - yold))         
    xold = xnew
    yold = ynew
    count =count +1 # What is the meaning of this line? It increments the iteration counter by 1
    # End of loop
    # Because we have two variables, we need to calculate the error for both and take the maximum
    # This is done in the e = max(....) line above
    #After the loop, we check if we reached the maximum number of iterations
    #The loop will terminate if either we reach the desired accuracy (e < eps) or if we reach the maximum number of iterations
    
if (count == maxIter):
    print(f"WARNING: Did not converge. e = {e:.4e}")

print(f"xnew = {xnew} ynew = {ynew} Error = {e:.4e} Count = {count}")
```
